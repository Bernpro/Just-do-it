```python
# Crear una tabla pivote
df_pivot = df.pivot_table(index='title', columns='user', values='rating').fillna(0)

# Convertir a matriz dispersa para eficiencia
from scipy.sparse import csr_matrix
matrix = csr_matrix(df_pivot.values)

# Configurar el modelo KNN
model = NearestNeighbors(metric='cosine', algorithm='brute')
model.fit(matrix)


def get_recommends(book = ""):
    # 1. Encontrar el índice del libro en la tabla pivote
    idx = df_pivot.index.get_loc(book)

    # 2. Buscar los 6 vecinos más cercanos (el primero es el libro mismo)
    distances, indices = model.kneighbors(df_pivot.iloc[idx].values.reshape(1, -1), n_neighbors=6)

    # 3. Formatear la salida según el requerimiento
    recommendations = []
    # Invertimos el orden para que coincida con el test de freeCodeCamp
    for i in range(len(distances[0])-1, 0, -1):
        recommendations.append([df_pivot.index[indices[0][i]], distances[0][i]])

    return [book, recommendations]
```
